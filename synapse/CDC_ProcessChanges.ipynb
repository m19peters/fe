```python
changes_path = "/data_swamp/incoming/crews/new_changes/"
table_path = "data_lake/bronze/crews/"
init_path = "data_swamp/incoming/crews/init/"
primary_key = "WR_NO"
dest_table_name = "CREWSSSP_WORK_REQUEST"
config_name = "crews"
init_cdc = 6275191283894
partition_by = "ENTRY_DATE_YEAR=YEAR(CAST(ENTRY_DATE* AS DATE)), ENTRY_DATE_MONTH=MONTH(CAST(ENTRY_DATE* AS DATE))"
```

```python
%%pyspark
import os
import pandas as pd

files = []

for file in mssparkutils.fs.ls(changes_path):
    ts = file.name[:14]

    if f"{dest_table_name}.txt" in file.name and str.isnumeric(ts):
        files.append(
        {
            'ts': int(ts),
            'file_name': file.name,
            'file_path': file.path
        })
    
df_files = pd.DataFrame(files).sort_values("ts", ascending=True)

for index, row in df_files.iterrows():
    print(row["ts"], row["file_name"], row["file_path"])
    schema_file =  f"{ row['ts'] }_schema.ini"
    change_file = row["file_name"]
    exitVal = mssparkutils.notebook.run(path="CDC", timeout_seconds=1200, arguments={
                                                            "changes_path" : changes_path,
                                                            "change_file" : change_file,
                                                            "schema_file" : schema_file,
                                                            "table_path" : table_path,
                                                            "init_path" : init_path,
                                                            "primary_key" : primary_key,
                                                            "dest_table_name" : dest_table_name,
                                                            "config_name" : config_name,
                                                            "init_cdc" : init_cdc,
                                                            "partition_by" : partition_by
                                                            })
    
    mv_path = row["file_path"].replace("/new_changes/", "/processed_changes/")
    print(f"finished processing {row['file_name']}.  Moving to {mv_path}")
    mssparkutils.fs.mv(row["file_path"], mv_path, True)
```
